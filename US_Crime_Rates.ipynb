# U.S. CRIME RATES BETWEEN 1964-2014

We are going to take a look at some data that contains US crime rates between 1960 - 2014
and see if you can learn something from it.
- Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

```python
import numpy as np
import pandas as pd
```

```python
# extracting the data and assign it to a variable called crime.
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')

# this piece of line will show the heading of the code.
crime.head()
```

We would like to check what type of info does each column contain ( is it a string, integer, boolean, etc):

```python
crime.info()
```

Based on the information above we know that each column (12 in total) contains non-empty integers.
Additionally, the index of our table goes from 0 to 54.

We would like to know not only the year, but also the exact day and month. Since each entry in the column 'Year'
is entered as some integer. We would have to change each entry to a type where we can get Year-Month-Day.

```python
crime.Year = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

Now Lets use the year as the new index for our table:

```python
crime.set_index('Year',inplace= True)
crime.head()
```

We are Going to remove the total per year in order to make an analysis of each decade:

```python
crime.drop('Total',axis=1,inplace= True)
crime.head()
```

In order to modify our table so we can obtain values per each decade, we must do the following:
    - Resample in order to sum each decade (except the population column)
    - Make sure we take the maximum value for the population column ONLY (per decade)

```python
# To learn more about .resample (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html)
# To learn more about Offset Aliases (http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases)

# Uses resample to sum each decade
crimes = crime.resample('10AS').sum()

# Uses resample to get the max value only for the "Population" column
population = crime['Population'].resample('10AS').max()

# Updating the "Population" column
crimes['Population'] = population

crimes
```

Since we dont have rates for the Year 2020 yet, we are going to filter (for now) the row 2020 from
the year column

```python
Temp_crimes = crimes[crimes['Population']>0]
Temp_crimes
```

We Would like to check which decade was the most and the least dangerous to live in the US:
- According to our results below, it seems like the US had high crime rates during the 90s

```python
Temp_crimes.idxmax(0)
```

 - Least dangerous: From our results below, we can estimate that the amounts of crime in the US was low during the 60s
     and possibly during the 2010s.

```python
Temp_crimes.idxmin(0)
```
